```python
%matplotlib inline
from pycocotools.coco import COCO
import numpy as np
import skimage.io as io
import matplotlib.pyplot as plt
import pylab
import json
pylab.rcParams['figure.figsize'] = (16, 16)
import json
```

```python
dataDir = "data/tiny_set/"
annFile = 'data/tiny_set/erase_with_uncertain_dataset/annotations/tiny_set_test_nobox.json'
outFile = 'data/tiny_set/erase_with_uncertain_dataset/annotations/tiny_set_test_nobox_cropped.json'
```

```python
with open(annFile) as f:
    test_set = json.load(f)
```

```python
new_images = []
```

```python
def crop_img(orig_w, orig_h, h, w):
    new_dims = []
    for i in range(0, orig_h, h):
        for j in range(0, orig_w, w):
            y = i
            x = j
            if y + h > orig_h:
                y = orig_h - h
            if x + w > orig_w:
                x = orig_w - w
            new_dims.append((x, y))
    return new_dims
```

```python
crop_img(2160, 3840, 640, 512)
```

```python
new_images = []
```

```python
i = 0
w = 512
h = 640
for img in test_set['images']:
    orig_w = int(img['width'])
    orig_h = int(img['height'])
    new_dims = crop_img(orig_h, orig_w, h, w)
    for dim in new_dims:
        x, y = dim
        new_img = img.copy()
        new_img['width'] = w
        new_img['height'] = h
        new_img['corner'] = [x, y, x+w, y+h]
        new_img['image_id'] = new_img['id']
        new_img['id'] = i
        new_images.append(new_img)
        i = i + 1
```

```python
new_json = {
    'type': 'instance',
    'images': new_images,
    'categories': [{'id': 1, 'name': 'person', 'supercategory': 'none'}]
}
```

```python
with open(outFile, 'w') as f:
    json.dump(new_json, f)
```
